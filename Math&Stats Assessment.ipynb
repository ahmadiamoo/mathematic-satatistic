```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("assessment_da25.csv")

# 1. Understanding the Data
print(data.info())
print(data.describe())

# 2. Explore Variants
sns.countplot(x='Variant', data=data)
plt.title('Distribution of Variants')
plt.show()

# 3. User Engagement
sns.scatterplot(x='Time on Page (sec)', y='Clicks on media', data=data)
plt.title('User Engagement Analysis')
plt.show()

# 4. Conversion Metrics
sns.barplot(x='Number of add to cart', y='GMV (in $)', data=data)
plt.title('Conversion Metrics Analysis')
plt.show()

# 5. GMV Analysis
sns.histplot(data['GMV (in $)'], bins=20, kde=True)
plt.title('Distribution of GMV')
plt.show()
```
